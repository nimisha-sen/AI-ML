## Real Time Detection

```python
cap = cv2.VideoCapture(0)
color_space_normal = True

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    def toGray() -> any:
        dst = np.zeros_like(src, dtype=np.uint8)  # Initialize destination matrix
    
        # Convert to grayscale
        gray_color_results = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
        
        # Perform Canny edge detection
        edges = cv2.Canny(gray_color_results, 50, 90)
        
        # Invert the edges
        dst = cv2.bitwise_not(edges)
        
        # Convert edges back to RGBA
        dst = cv2.cvtColor(dst, cv2.COLOR_GRAY2RGBA)
        
        # Resize dst to match frame dimensions
        dst = cv2.resize(dst, (frame.shape[1], frame.shape[0]), interpolation=cv2.INTER_AREA)
        
        return dst
    
    # Make detections 
    results = model(frame)
    
    src = np.squeeze(results.render())

    key = cv2.waitKey(10)
    
    # Setting it first time
    if color_space_normal:
        cv2.imshow('YOLO', src)
    elif color_space_normal == False:
        cv2.imshow('YOLO', toGray())
    
    # This is for settinging it while its running
    if key & 0xFF == ord('t'):
        color_space_normal = not (color_space_normal) # Invert
        if color_space_normal:
            cv2.imshow('YOLO', src)
        elif color_space_normal == False:
            cv2.imshow('YOLO', toGray())
    elif key & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()


```
